```python
pesticides = ["NeemOil", "Chlorpyrifos", "Malathion", "Imidacloprid", "Spinosad"]
effectiveness = [75, 85, 70, 90, 88]

# 1. Most effective pesticide
max_eff = max(effectiveness)
most_effective = pesticides[effectiveness.index(max_eff)]
print("1. Most effective pesticide:", most_effective, "-", max_eff, "%")

# 2. Least effective pesticide
min_eff = min(effectiveness)
least_effective = pesticides[effectiveness.index(min_eff)]
print("2. Least effective pesticide:", least_effective, "-", min_eff, "%")

# 3. Average effectiveness
average = sum(effectiveness) / len(effectiveness)
print("3. Average effectiveness:", average, "%")

# 4. Pesticides equal to or above average
print("4. Pesticides with effectiveness ≥ average:")
for i in range(len(effectiveness)):
    if effectiveness[i] >= average:
        print(pesticides[i], "-", effectiveness[i], "%")

# 5. Pesticides above 80%
print("5. Pesticides with effectiveness above 80%:")
for i in range(len(effectiveness)):
    if effectiveness[i] > 80:
        print(pesticides[i], "-", effectiveness[i], "%")

# 6. Count pesticides below 80%
count_below_80 = 0
for eff in effectiveness:
    if eff < 80:
        count_below_80 += 1
print("6. Number of pesticides below 80%:", count_below_80)

# 7. Increase effectiveness by 5%
updated_effectiveness = []
for eff in effectiveness:
    updated_effectiveness.append(eff + 5)
print("7. Updated effectiveness (+5%):", updated_effectiveness)

# 8. Suitable for organic farming (≥ 85%)
print("8. Pesticides suitable for organic farming:")
for i in range(len(effectiveness)):
    if effectiveness[i] >= 85:
        print(pesticides[i], "-", effectiveness[i], "%")

# 9. Replace pesticides below 75% as 'Not Recommended'
recommended_list = []
for i in range(len(effectiveness)):
    if effectiveness[i] < 75:
        recommended_list.append("Not Recommended")
    else:
        recommended_list.append(pesticides[i])
print("9. Recommendation list:", recommended_list)

# 10. Sort pesticides based on effectiveness (ascending)
sorted_data = sorted(zip(effectiveness, pesticides))
print("10. Pesticides sorted by effectiveness (ascending):")
for eff, pest in sorted_data:
    print(pest, "-", eff, "%")
```

```python
# Farm Equipment Cost Analysis

equipment = ["Tractor", "Plough", "Harvester", "Seeder", "Sprayer", "Thresher"]
cost = [850000, 25000, 1200000, 60000, 18000, 95000]

# 1. Most expensive equipment
max_cost = max(cost)
print("1.Most expensive equipment:", equipment[cost.index(max_cost)], "-", max_cost)

# 2. Least expensive equipment
min_cost = min(cost)
print("2.Least expensive equipment:", equipment[cost.index(min_cost)], "-", min_cost)

# 3. Difference between most and least expensive
print("3.Cost difference:", max_cost - min_cost)

# 4. Average cost
avg_cost = sum(cost) / len(cost)
print("4.Average cost:", avg_cost)

# 5. Equipment equal to average cost
print("5.Equipment with average co mst:")
found = False
for i in range(len(cost)):
    if cost[i] == avg_cost:
        print(equipment[i])
        found = True
if not found:
    print("None")

# 6. Equipment costing more than average
print("6.Equipment costing more than average:")
for i in range(len(cost)):
    if cost[i] > avg_cost:
        print(equipment[i], cost[i])

# 7. Count equipment costing less than 50,000
count_50k = 0
for c in cost:
    if c < 50000:
        count_50k += 1
print("7.Items costing less than 50,000:", count_50k)

# 8. Equipment eligible for subsidy (≤ 1,00,000)
print("8.Equipment eligible for subsidy:")
for i in range(len(cost)):
    if cost[i] <= 100000:
        print(equipment[i], cost[i])

# 9. Equipment requiring loan support (≥ 5,00,000)
print("9.Equipment requiring loan support:")
for i in range(len(cost)):
    if cost[i] >= 500000:
        print(equipment[i], cost[i])

# 10. Equipment suitable for small-scale farmers (< 75,000)
print("10.Small-scale farmer equipment:")
for i in range(len(cost)):
    if cost[i] < 75000:
        print(equipment[i], cost[i])

# 11. Sort equipment based on cost (ascending)
sorted_data = sorted(zip(cost, equipment))
print("11.Sorted equipment by cost:")
for c, e in sorted_data:
    print(e, c)

# 12. Total cost
print("12.Total cost of all equipment:", sum(cost))

# 13. Count equipment costing between 50,000 and 1,00,000
count_range = 0
for c in cost:
    if 50000 <= c <= 100000:
        count_range += 1
print("13.Items costing between 50,000 and 1,00,000:", count_range)

# 14. Increase cost by 8%
updated_cost = []
for c in cost:
    updated_cost.append(c * 1.08)
print("14.Cost after 8% increase:", updated_cost)

# 15. Replace cost above 10,00,000 as "High Investment"
investment_category = []
for c in cost:
    if c > 1000000:
        investment_category.append("High Investment")
    else:
        investment_category.append(c)
print("15.Investment category list:", investment_category)

# 16. Equipment costing more than twice the minimum cost
print("16.Cost more than twice the minimum cost:")
for i in range(len(cost)):
    if cost[i] > 2 * min_cost:
        print(equipment[i], cost[i])




# 17. Check if all costs are below 15,00,000
all_below = True
for c in cost:
    if c >= 1500000:
        all_below = False
print("17.All costs below 15,00,000:", all_below)




# 18. Top three most expensive equipment
sorted_desc = sorted(zip(cost, equipment), reverse=True)
print("18.Top three expensive equipment:")
for i in range(3):
    print(sorted_desc[i][1], sorted_desc[i][0])



# 19. Reduce subsidy equipment cost by 15%
reduced_cost = []
for c in cost:
    if c <= 100000:
        reduced_cost.append(c * 0.85)
    else:
        reduced_cost.append(c)
print("19.Cost after subsidy reduction:", reduced_cost)




# 20. Cost categories (Low, Medium, High)
categories = []
for c in cost:
    if c < 50000:
        categories.append("Low")
    elif c <= 500000:
        categories.append("Medium")
    else:
        categories.append("High")

print("20.Cost categories:")
for i in range(len(equipment)):
    print(equipment[i], "-", categories[i])
```
