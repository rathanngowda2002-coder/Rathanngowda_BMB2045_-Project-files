````python
import pandas as pd
import numpy as np
data=pd.DataFrame({
    'plot':['A','B','C','D','E'],
    'yield_kg':[2500,np.nan,2800,np.nan,2600],
    'rainfall':[150,200,np.nan,180,170],
    'temperature':[28,30,29,np.nan,27]
})
print("===Data with missing values===")
print(data)
````

````python
print('Data with missig values')
print(data)
print('Detecting missing values')
print('Is null?')
print(data.isnull())
print('Total missing column')
print(data.isnull().sum())
````

````python
print('Filling strategies')
data_mean=data.copy()
print(data_mean['yield_kg'])
data_mean['yield_kg']=data_mean['yield_kg'].fillna(data_mean['yield_kg'].mean())
print('fill with mean')
print(data_mean['yield_kg'])
````

````python
data_median=data.copy()
print(data_median['rainfall'])
data_median['rainfall']=data_median['rainfall'].fillna(data_median['rainfall'].median())
print(data_median['rainfall'])
````

````python
data_ffill=data.copy()
print(data_ffill)
data_ffill=data_ffill.fillna(method='ffill')
print("Forward fill:")
print(data_ffill)
````

````python
print('Dropping rows')
data_dropped=data.dropna()
print(data_dropped)
print('original rows:{len(data)},After drop:{len(data_dropped)}')
````

````python
import pandas as pd
import io

csv_data = """
Years Experience,Salary
1.1,39343
1.3,46205
1.5,37731
2,43525
2.2,39891
2.9,56642
3,60150
3.2,54445
3.2,64445
3.7,57189
3.9,63218
4,55794
4,56957
4.1,57081
4.5,61111
4.9,67938
5.1,66029
5.3,83088
5.9,81363
6,93940
6.8,91738
7.1,98273
7.9,101302
8.2,113812
8.7,109431
9,105582
9.5,116969
9.6,112635
10.3,122391
10.5,121872
"""
dataset = pd.read_csv(io.StringIO(csv_data))
print(dataset.head())
print(dataset.tail())
````

slice the dataset as descriptive(independent) and target variable(dependent)

````python
X=dataset.iloc[:,0:1].values
Y=dataset.iloc[:,-1].values
print(X)
print(Y)
````

````python
#splitting dataset into training set and testing set
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=1/3,random_state=0)
print(dataset.shape)
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
````

````python
#calculate the mean values of input and output
x_mean=np.mean(X)
y_mean=np.mean(Y)
print(X)
print(Y)
print(x_mean)
print(y_mean)
````

````python
input_mean=np.mean(X)
output_mean=np.mean(Y)
print(X)
print(Y)
print(input_mean)
print(output_mean)
````

````python
 n=len(X_train)
 num,den=0,0
 for i in range(n):
  num+=(X_train[i]-x_mean)*(Y_train[i]-y_mean)
  den+=(X_train[i]-x_mean)**2
b1=num/den
b0=y_mean-(b1*x_mean)
print("b0",b0)
print("b1",b1)
````

````python
y_pred_code=b0+b1*X_test
print(y_pred_code)
````

````python
print(Y_test)
````

````python
#mean square error
n1=len(Y_test)
mse=0
for i in range(n1):
  mse+=(Y_test[i]-y_pred_code[i])**2
  mse=mse/n1
print(mse)
rmse=np.sqrt(mse)
print(rmse)
````

Machine learning algorithm

````python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
````

````python
reg.fit(X_train,Y_train)
````

````python
y_pred=reg.predict(X_test)
print(y_pred)
````

````python
print('Intercept',reg.intercept_)
print('coeficient',reg.coef_)
````

Evaluation metrics

````python
from sklearn import metrics
print("Mean absolute error:")
print(metrics.mean_absolute_error(Y_test,y_pred))
````

````python
print("Mean absolute error:",metrics.mean_absolute_error(Y_test,y_pred))
````

````python
print("Mean squared error:",metrics.mean_squared_error(Y_test,y_pred))
````

````python
print("Root mean square error:",np.sqrt(metrics.mean_squared_error(Y_test,y_pred)))
````
