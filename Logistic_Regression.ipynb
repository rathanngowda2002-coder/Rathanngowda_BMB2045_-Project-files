```python
#read the dataset
import pandas as pd
data=pd.read_csv("iris.csv")
print(data)
```

```python
data.info()
```

```python
data["Species"].unique()
```

```python
data["Species"].value_counts()
```

```python
#separate dependent and independent variables(input/output)
x=data.iloc[:,:-1].values
y=data.iloc[:,4].values
print(x)
```

```python
#split the dataset into train and test
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=1/3,random_state=0)
```

```python
#create decision tree object
from sklearn.tree import DecisionTreeClassifier
classifier=DecisionTreeClassifier(criterion='entropy',max_depth=None,min_samples_split=5,min_samples_leaf=2)
```

```python
#train model('fit' method)
model = classifier.fit(x_train,y_train)
print(model)
```

```python
#predict the testing model
y_pred=model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix
c=confusion_matrix(y_test,y_pred)
c
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_pred))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python
#visualisation
from sklearn import tree
fn=['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)']
cn=['setosa','veresicolor','virginica']
fig,axes=plt.subplots(nrows=1,ncols=1,figsize=(4,4),dpi=300)
tree.plot_tree(model,
               feature_names=fn,
               class_names=cn,
               filled=True)
```
