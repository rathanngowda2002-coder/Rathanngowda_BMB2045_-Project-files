```python
import pandas as pd
df=pd.read_csv("agri_data.csv")
#pd.set_option('display.expand_frame_repr',false)
print(df)
```

```python
#filter data(only wheat)
wheat_data=df[df["Crop"]=="Wheat"]
print("\nwheat_data:")
print(wheat_data)
```

```python
#group by year and sum production
yearly_production=df.groupby("Year")["Production_Tons"].sum()
print("\nyearly total production:")
print(yearly_production)
```

```python
import matplotlib.pyplot as plt
#create bar graph
plt.figure(figsize=(8,5))
bars = plt.bar(yearly_production.index,yearly_production.values,color="skyblue")
```

```python
plt.xlabel("Year")
plt.ylabel("Total area(hectares)")
plt.title("Total Production by Year")
plt.show()
```

```python
#bar plot-production by crop(single year example:2022)
#filter yeat 2022
data_2022=df[df["Year"]==2022]
print(data_2022)
```
