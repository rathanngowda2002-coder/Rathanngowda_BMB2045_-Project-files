```python
#create data with missing values
import pandas as pd
import numpy as np
data=pd.DataFrame({
    "plot:":['A','B','C','D','E'],
    "yield_kg":[2500,np.nan,2800,np.nan,2600],
    "rainfall":[150,280,np.nan,180,170],
    "temperature":[28,30,29,np.nan,27]
})
print("==Data with missing values==")
print(data)

#==detecting missing data
print("==Detecting missing data==")
print(data.isnull())




print(data.isnull().sum())
```

```python
#fill with mean
data_mean=data.copy()
data_mean["yield_kg"].fillna(data_mean["yield_kg"].mean(),inplace=True)
print('\n','fill with mean')
print(data_mean["yield_kg"])
```

```python
#fill the median
data_median=data.copy()
print(data_median['rainfall'])
data_median['rainfall'].fillna(data_median['rainfall'].median(),inplace=True)
print('\n','fill with median')
print(data_median['rainfall'])
```

```python
#forward fill(use previous value)
data_ffill=data.copy()
print(data_ffill)
data_ffill=data_ffill.fillna(method='ffill')
print('\n','forward fill')
print(data_ffill)
```

```python
#==DROPPING MISSING DATA
print("\n== Dropping Rows ===")

data_dropped = data.dropna()
print(data_dropped)
print(f"original rows:{len(data)}")
```

```python
import pandas as pd
df=pd.read_csv("Salary_Data.csv")

#slice the dataset as descriptive (independent)
#target variable(dependent)
x=dataset.iloc[:,0:1].values
y=dataset.iloc[:,-1].values
print(x)
print(y)
```

```python
#splitting the dataset into training set and testing set
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
print(dataset.shape)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
print("data split successfully")
```

```python
import numpy as np

#calculate the mean values of input and output
x_mean=np.mean(x)
y_mean=np.mean(y)
print(x_mean)
print(y_mean)
```

```python
#calculate intercept and coefficient
n=len(x)
num,den=0,0
for i in range(n):
  num+=(x[i]-x_mean)*(y[i]-y_mean)
  den+=(x[i]-x_mean)**2

b1=num/den
b0=y_mean - (b1*x_mean)
print("b0",b0)
print("b1",b1)
```

```python
#predicting results on test data
y_pred_code=b0+b1*x_test
print(y_pred_code)
```

```python
print(y_test)
```

```python
#mean square error
n1=len(y_test)
mse=0
for i in range(n1):
  mse+=(y_pred_code[i]-y_test[i])**2
  mse=mse/n1
print(mse)
rmse=np.sqrt(mse)
print(mse)
```

```python
#create an object of the algorithm/model
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
```

```python
#predict the model on testing dataset
y_pred=reg.predict(x_test)
print(y_pred)
```

```python
#evaluation metrics
from sklearn import metrics
print("mean absolute error:")
print(metrics.mean_absolute_error(y_test,y_pred))
```

```python
print("mean square error:",metrics.mean_squared_error(y_test, y_pred))
```

```python
print("root mrsn square error:",np.sqrt(metrics.mean_squared_error(y_test,y_pred)))
```
