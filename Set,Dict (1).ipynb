```python
#factorial of a number
def fact(m):
  fact_result = 1
  if(m > 0):
   for i in range(1, m + 1):
    fact_result = fact_result * i
   return fact_result
  elif m == 0:
    return 1
  else:
    return "Factorial is not defined for negative numbers"
m = int(input("enter a number"))
factorial_value = fact(m)
print("factorial of", m, "is", factorial_value)
```

```python
import pandas as pd
data={
    "farm":["farm a","farm b","farm c","farm d"],
    "crop":["wheat","rice","wheat","maize"],
    "area_hectares":[10,15,8,12],
    "production_tons":[30,60,24,36]
}
df=pd.DataFrame(data)
print(df)
```

```python
df.head(2)
```

```python
#yield=production/area
df["yield_ton_per_hectare"]=df["production_tons"]/df["area_hectares"]
print(df)
```

```python
df.iloc[[1,3]]
```

```python
df.iloc[[1,3],[0,1,2]]
```

```python
#filter data
#farms growing wheat
print(df["crop"]=="wheat")
print(df)
wheat_farms=df[df["crop"]=="wheat"]
print(wheat_farms)
```

```python
#sort the farms by highest production
sorted_df = df.sort_values(by="production_tons",ascending=False)
print(sorted_df)
```

```python
#groupby operation  - average yield per crop
avg_yield = df.groupby("crop")["yield_ton_per_hectare"].mean()
print(avg_yield)
```

```python
#groupby operation  - average yield per crop
avg_yield = df.groupby("crop")["yield_ton_per_hectare"].sum()
print(avg_yield)
```
